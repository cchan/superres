```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
import glob

images = [np.flip(cv2.imread(filename), axis=2) for filename in glob.glob("img/*.png")]

index = np.random.choice(len(images))
print('index:', index)

image = images[index]
print('image shape:', np.shape(image))

plt.figure()
plt.imshow(image)
plt.show()

plt.figure()
plt.imshow(cv2.resize(image, dsize=(0,0), fx=0.2, fy=0.2, interpolation=cv2.INTER_CUBIC))
plt.show()
```
