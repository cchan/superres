```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
import glob

np.random.seed(24601)
factor = 0.25

images = [np.flip(cv2.imread(filename), axis=2) for filename in glob.glob("img/*.png")]
imagessmall = [cv2.resize(image, dsize=(0,0), fx=factor, fy=factor, interpolation=cv2.INTER_CUBIC) for image in images]

i = np.random.choice(len(images))
print('index:', i)
print('image shape:', np.shape(images[i]))
print('imagessmall shape:', np.shape(imagessmall[i]))

plt.figure()
plt.imshow(images[i])
plt.show()

plt.figure()
plt.imshow(imagessmall[i])
plt.show()
```
